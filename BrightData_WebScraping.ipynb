**Jupyter notebooks, including Google Colab**, operate within an environment where an event loop is already running. This environment is designed to handle interactive tasks, such as running cells, displaying outputs, and handling user inputs concurrently.



The asyncio.run() function cannot be used in these environments because it tries to create and run a new event loop, leading to a RuntimeError


**nest_asyncio** is a library that allows nested usage of the asyncio event loop, effectively enabling the running of new asynchronous tasks within an already running event loop. This is crucial for making asyncio work in Jupyter notebooks and Colab environments.
